```python
import requests
import pandas as pd
import time

def get_lat_lon(address, rest_api_key):
    url = "https://dapi.kakao.com/v2/local/search/address.json"
    headers = {"Authorization": f"KakaoAK {rest_api_key}"}
    params = {"query": address}

    try:
        response = requests.get(url, headers=headers, params=params)
        result = response.json()
        if result.get('documents'):
            match = result['documents'][0]
            return float(match['y']), float(match['x'])
        return None, None
    except:
        return None, None

KAKAO_API_KEY = "158ffd3096749420147ae17e585d274d"
file_path = "물리적리스크.xlsx"

df = pd.read_excel(file_path)

latitudes = []
longitudes = []

for idx, row in df.iterrows():
    if pd.isna(row['사업장 위치']):
        latitudes.append(None)
        longitudes.append(None)
        continue

    address = str(row['사업장 위치'])
    lat, lon = get_lat_lon(address, KAKAO_API_KEY)

    latitudes.append(lat)
    longitudes.append(lon)
    time.sleep(0.1)

df['위도(Lat)'] = latitudes
df['경도(Lon)'] = longitudes

df.to_excel("위경도_변환완료.xlsx", index=False)
```
